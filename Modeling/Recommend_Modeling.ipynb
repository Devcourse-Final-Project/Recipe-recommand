```python
def split_value(df):
    data = []
    for i in range(len(df)):
        cv = []
        if isinstance(df.loc[i, 'food_name'], str):
            k = df.loc[i, 'food_name'].split('  ')
            cv.extend(k)
            if len(k) < 30:
                cv.extend([np.nan] * (30-len(k)))
        else:
            cv.extend([np.nan]*30)
        data.append(cv)
    category_df = pd.DataFrame(data, columns = ['food1', 'food2', 'food3', 'food4', 'food5', 'food6', 'food7', 'food8', 'food9', 'food10', 'food11', 'food12', 'food13', 'food14', 'food15', 'food16', 'food17', 'food18', 'food19', 'food20', 'food21', 'food22', 'food23', 'food24', 'food25', 'food26', 'food27', 'food28', 'food29', 'food30'])
    return category_df
```

```python
# 날짜 변환
def convert_date(df):
    cre_date = df['created_date'].tolist()
    for i in range(len(cre_date)):
        if '.' in cre_date[i]:
            cre_date[i] = cre_date[i].replace('.', '').replace(' ', '-')
        elif '실시간-스트리밍-시작일:' in cre_date[i]:
            cre_date[i] = cre_date[i].replace('실시간-스트리밍-시작일:-', '')
        elif cre_date[i] is np.nan:
            continue
        else:
            continue
    df['created_date'] = cre_date
    return df
```

```python
from sklearn.neighbors import NearestNeighbors

nbrs = NearestNeighbors(n_neighbors=11).fit(category_k)
nbrs
```

```python
data = []
for i in range(len(category_k)):
    ck = category_k.iloc[i, :]
    distances, indexes = nbrs.kneighbors([ck])
    recom = grouped.loc[indexes[0], ['food_name']]
    recom['distance'] = distances[0]
    rec = recom['food_name'].iloc[-10:].to_list()
    data.append(rec)
```
